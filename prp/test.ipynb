```python
# Removes the duplicate rows
df.drop_duplicates()
```

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
if 'stock_level' in df.columns:
    q1 = df['stock_level'].quantile(0.25)
    q3 = df['stock_level'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df['stock_level'] >= lower_bound) & (df['stock_level'] <= upper_bound)]
```

```python
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print("\n🔹 Checking Missing Values After Handling:")
print(df.isnull().sum())
```

```python
print("Number of Duplicate Rows Before Removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Number of Duplicate Rows After Removal:", df.duplicated().sum())
```

```python
print("Data Types Before Correction:")
print(df.dtypes)
```

```python
date_cols = ['Date', 'last_updated']  
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
print("Data Types After Correction:")
print(df.dtypes)
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python

if 'stock_level' in df.columns:
    print("Stock Level Outliers Before Removal (IQR):", df.shape[0])
    df = remove_outliers_iqr(df, 'stock_level')
    print("Stock Level Outliers After Removal (IQR):", df.shape[0])
```

```python
if 'stock_level' in df.columns:
    df['stock_status'] = np.where(df['stock_level'] < 10, 'Low Stock', 'Sufficient Stock')
```

```python
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.histplot(df['stock_level'], bins=20, kde=True)
plt.title("Stock Level Distribution")
plt.show()
```
