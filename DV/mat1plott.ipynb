```python
x=["Jan","Feb","March","April","May","June"]
y=[1000,500,8000,1100,500,6000]
plt.plot(x,y,linestyle='dashed',linewidth=2,marker='o',color='green')
plt.title("6 Month Growth Saving Account " )
plt.xlabel("Months")
plt.ylabel("Savings")
plt.show()
```

To use =====Bar Chart=========  just plt.plot -> plt.bar

```python
x=["Jan","Feb","March","April","May","June"]
y=[1000,500,8000,1100,500,6000]
plt.bar(  x,y,color='yellow',width=.3)
#plt.bar(x is categorries,y is values, is color )

plt.title("6 Month Growth Saving Account " )
plt.xlabel("Months")
plt.ylabel("Savings")
plt.show()
```

======For different colors========

```python
plt.bar(  x,y,color=['yellow','green','black','gray'],width=.7)
```

============edgecolor='violet',==================

```python
plt.bar(  x,y,color=['yellow','green','black','gray'],edgecolor='violet',width=.7)
plt.title("6 Month Growth Saving Account " )
```

+++++++++++++++++++ Histogram =======================

```python
ages = [5,20,22,22,23,25,27,27,28,30,31,32,34,35,35,36,45,63]
b=[1,20,30,40,50,60,70]
plt.hist(ages,bins=b,color='red',edgecolor='green')
plt.title("Age Distribution of Customers")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.grid(True) # To get grid values
plt.show()
```

===>Piechart

```python
subs =["java","Python","C","C++","Ruby"]
sizes=[10,30,40,5,15]

colors=["green","black","yellow","violet","gray"]
plt.pie(sizes,labels=subs,autopct='%1.1f%%', colors=colors,startangle=90 )
# start angle is start of the roatte of chart to 90 degrees
plt.title("Programming Language Usage")


plt.show()
```

================== Scatter plot========

```python
x = [5,10,15,20,25]
y = [10,20,25,30,35]

plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot Examples")
plt.show()
```
