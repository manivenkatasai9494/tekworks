```python
import pandas as pd

# Sample data (replace with your actual data loading step)
data = {
    'Company': ['Yerevan Brandy Company'],
    'Date': ['27 January 2004, 18:00']
}

# Creating a DataFrame (replace with your actual DataFrame loading)
df = pd.DataFrame(data)

# Step 1: Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)

# Step 2: Extract new features from 'Date'
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.weekday  # Monday=0, Sunday=6
df['Quarter'] = df['Date'].dt.quarter

# Step 3: Create a 'Weekend' column (1 for Saturday/Sunday, else 0)
df['Weekend'] = df['Weekday'].apply(lambda x: 1 if x >= 5 else 0)

# Step 4: Verify the extracted features
print(df[['Date', 'Year', 'Month', 'Day', 'Weekday', 'Quarter', 'Weekend']])

# Optionally, you can save the result to a CSV
# df.to_csv('processed_job_postings.csv', index=False)
```
