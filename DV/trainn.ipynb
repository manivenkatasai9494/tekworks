```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
sns.barplot(data=df,x='Pclass',y='Survived', errorbar=None )
```

```python
sns.barplot(data=df,x='Pclass',y='Survived', errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age' ,y='Fare'  , hue='Survived',palette='coolwarm',alpha=0.6)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.show()
```

```python

sns.scatterplot(x='Age', y='Pclass', data=df)
plt.title('Age vs Pclass')
plt.show()

```

```python
sns.scatterplot(x='Fare', y='Pclass', data=df)
plt.title('Fare vs Pclass')
plt.show()
```
