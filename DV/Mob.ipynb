#### 2️ String & Categorical Data Preprocessing
    Convert categorical data into lowercase and remove special characters (make,
    fuel_type, aspiration, etc.).
    Replace "?" or NaN values with appropriate replacements.
    Standardize number_of_cylinders:
    Convert words ("four", "six", "eight") to integers (4, 6, 8).
    Standardize fuel_system categories (e.g., mpfi, spdi → Multi-Point Fuel Injection,
    Single-Point Fuel Injection).

```python
for i in auto_df:
    if auto_df[i].dtype=='object':
        auto_df[i]=auto_df[i].str.lower()
        auto_df[i]=auto_df[i].str.strip()
        
```

#### 3️⃣Numeric Feature Processing
    Convert "horsepower" and "peak_rpm" to integers (handle missing values first).
    Convert price to numeric and fill missing values.
    Standardize units (e.g., convert horsepower to kW if needed).


```python
auto_df['horsepower'] = auto_df['horsepower'].astype(int)
```

### 4.Feature Engineering on Numeric Columns
    Create New Features:
    Power-to-weight ratio = horsepower / curb_weight.
    Engine efficiency = horsepower / engine_size.
    Fuel efficiency = city_mpg / highway_mpg.
    Convert compression_ratio into categories:
    Low (<9), Medium (9-11), High (>11).

```python
auto_df['power_to_weight'] = auto_df['horsepower']/auto_df['curb_weight']
```

```python
auto_df['engine_efficiency'] =auto_df['horsepower']/auto_df['engine_size']
```

```python
auto_df['fuel_efficiency'] = auto_df['city_mpg']/auto_df['highway_mpg']
```

```python
def convert_compress(a):
    if a<9:
        return 'Low'
    elif a>9 and a<11:
        return 'Medium'
    else:
        return 'High'
```

```python
auto_df['compression_ratio'] = auto_df['compression_ratio'].apply(convert_compress)
```

```python
auto_df['compression_ratio']
```

### 5️⃣ Handling Date or Range Values
    Extract ranges from normalized_losses and create bins (e.g., Low, Medium, High
    loss categories).

```python
auto_df['normalized_losses'].describe()
```

```python
def normalised_(a):
    if a<101:
        return 'Low'
    elif a>101 and a<122:
        return 'Medium'
    else:
        return 'High'
```

```python
auto_df['normalized_losses']= auto_df['normalized_losses'].apply(normalised_)
```

### 6️⃣Encoding Categorical Features
    Label Encoding for fuel_type, aspiration, drive_wheels, etc.
    One-Hot Encoding for body_style, engine_type, and fuel_system.


```python
from sklearn.preprocessing import LabelEncoder
```

```python
body_style_dummies = pd.get_dummies(auto_df['body_style'])
```

```python
engine_type_dummies = pd.get_dummies(auto_df['engine_type'])
```

```python
fuel_system_dummies = pd.get_dummies(auto_df['fuel_type'])
```

```python
auto_df = pd.concat([auto_df,body_style_dummies,engine_type_dummies,fuel_system_dummies],axis=1)
```
