```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
print(tips.head(30))
```

```python
tips.shape
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="magenta")
plt.title("Distribution of Total Bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill", fill=True,color="yellow" )
plt.title("KDE Plot Of Total Bills")
plt.show()
```

<h2>Box plot </h2>

```python
sns.barplot(data=tips,x="day",y="total_bill",hue="viridis")
plt.title("Average Total Bill by Day")
plt.show()
```
