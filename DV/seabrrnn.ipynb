```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
print(tips.head(30))
```

```python
tips.shape
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="magenta")
plt.title("Distribution of Total Bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill", fill=True,color="yellow" )
plt.title("KDE Plot Of Total Bills")
plt.show()
```

<h2>Box plot </h2>

```python
sns.barplot(data=tips,x="day",y="total_bill",palette="viridis")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill", hue="sex" , palette="coolwarm" )
plt.title("Box plot total Bills by Day and Gender")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="green")
plt.title("Line plot of party size vs.Tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python
numtips = tips.select_dtypes(["number"])
correlation_matrix = numtips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
