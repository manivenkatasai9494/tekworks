```python
arr=np.arange(12).reshape(4,3)
print(np.sum(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
```

```python
print(np.sum(arr ,axis=1))   # axis is 0 means going to vertical line
print(np.sum(arr,axis=1))    # axis is 1 means going to horizontal axis like we manipulating at the horintal wise only
```

```python
arr=np.arange(12).reshape(3,4)
print(np.sum(arr,axis=0))
print(np.min(arr,axis=0))
print(np.max(arr,axis=0))
print(np.mean(arr,axis=0))

print("------------------------")
print(np.sum(arr,axis=1))
print(np.min(arr,axis=1))
print(np.max(arr,axis=1))
print(np.mean(arr,axis=1))
```
