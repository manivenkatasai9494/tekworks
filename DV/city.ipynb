```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv')
print(data)
```

```python
data.shape
```

```python
pd.unique(data['Country'])
```

```python
city_data = data[data['City'] == 'New York'].copy()
city_data
```

```python
data['dt'] =pd.to_datetime(data['dt'],errors='coerce')
```

```python
data.dropna
data['Year'] = data['dt'].dt.year
annual_avg_temp = data.groupby('Year')['AverageTemperature'].mean().reset_index()
```

```python
plt.Figure(figsize=(12,6))
plt.plot(annual_avg_temp['Year'] , annual_avg_temp['AverageTemperature'],marker='o')
```

Which Countries Had the highest Average temp in the Dataset

```python
data['Country'].head(10)
```

```python
unique_cities=pd.unique(data['Country'])[2:12]
unique_cities
```

```python
country_avg_temp = data.groupby('Country')['AverageTemperature'].mean().reset_index()
#country_avg_temp = data.groupby('Country')['Temperature'].mean().reset_index()
```

```python
country_avg_temp = country_avg_temp.dropna
```

```python
top_countries = country_avg_temp.sort_values(by='AverageTemperature', ascending=False).head(10)
print(top_countries)
plt.figure(figsize=(12,6))
plt.bar(top_countries[''])
```

```python
plt.figure(figsize=(12,6))
plt.bar(top_countries['Country'], top_countries['AverageTemperature']) 
plt.xlabel('Country')  
plt.ylabel('Average Temperature')  
plt.title('Top 10 Countries by Average Temperature')  
plt.xticks(rotation=45)  
plt.show()  
```
