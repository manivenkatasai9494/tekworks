<h1>Streaming
<h3>Streaming means you get the answer little by little,
instead of waiting for the full answer at the end.
Faster first response

Feels more interactive

Best for long answers

Used in chat apps (ChatGPT style)

<b><h1>without streaming</h1>

```python
from langchain_groq import ChatGroq
from langchain.messages import HumanMessage

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
messages = [
    HumanMessage(content="Why do parrots talk? in 600 words")
]
response = llm.invoke(messages)
print(response.content)
```

<h1> with streaming

```python
from langchain_groq import ChatGroq
from langchain.messages import HumanMessage
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
messages = [
    HumanMessage(content="Why do parrots talk in 900 words?")
]

print("Assistant: ", end="", flush=True) # flush show txt immediately
for chunk in llm.stream(messages):   # returns small chunks time to time
    if chunk.content:
        print(chunk.content, end="", flush=True)
```

<h1>Batch</h1>
Batching = sending MANY questions at ONE TIME

<h3>Type 1:</h3>
You give 3 questions

LangChain sends them together

It decides: (make cook all at once)

  How many run in parallel

  How fast they run

  You don’t control anything.
api problem limit exceeds

```python
responses = llm.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])

for response in responses:
    print(response.content)
```

<h1>Type 2 <h1>
You still send 5 questions together

But you say:

“Don’t run more than 5 at a time”

LangChain obeys your limit.

```python
llm.batch(
    ["Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"],
    config={
        'max_concurrency': 5,  # Limit to max 5 questions
    }
)
```

<b>2.Human message</b> - Represents user input and interactions with the model
<b>AI message </b>- Responses generated by the model, including text content, tool calls, and metadata <b>("Write a poem on artificial intelligence") </b>  <br>
<b>3.Tool message </b> - Represents the outputs of tool calls


== System + Human + AI history → Model → New answer ==

# 1.Pydantic
Pydantic models provide the richest feature set with field validation, descriptions, and nested structures.

```python
from pydantic import BaseModel,Field
# whenever i talk abt movie u should anser
class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="This year the movie was released")
    director:str=Field(description="The director of the movie")
    rating:float=Field(description="The movies rating out of 10")
```

```python
model_with_structure=model.with_structured_output(Movie)
```

# without pydantic

```python
model.invoke("Provide details about the moview Bahubali")
```

# with pydantic

```python
response=model_with_structure.invoke("Provide details about the movie bahubali")
response
```

```python
response=model_with_structure.invoke("Provide details about the movie squit game 2")
response
response=model_with_structure.invoke("Provide details about the movie stranger things 5")
response
```

# Nested Structure

```python
from pydantic import BaseModel, Field

class Actor(BaseModel):
    name: str
    role: str

class MovieDetails(BaseModel):
    title: str
    year: int
    cast: list[Actor]
    genres: list[str]
    budget: float | None = Field(None, description="Budget in millions USD")

model_with_structure = model.with_structured_output(MovieDetails)

response = model_with_structure.invoke("Provide details about the movie bahubali")
response
```


# TypedDict
TypedDict is like giving the AI a fixed form to fill,
so the answer always comes in a clean, structured format <br>

Imagine you go to a movie theatre office and ask:
<br>
“Give me Avengers movie details”
<br>
If you don’t give a format, staff may talk freely ❌ <br>
If you give a form, they must fill it properly ✅


```python
# schema created
from typing_extensions import TypedDict,Annotated

class MovieDict(TypedDict):
    """A movie with details."""
    title: Annotated[str, ..., "The title of the movie"]
    year: Annotated[int, ..., "The year the movie was released"]
    director: Annotated[str, ..., "The director of the movie"]
    rating: Annotated[float, ..., "The movie's rating out of 10"]


model_withtypedict=model.with_structured_output(MovieDict)
response=model_withtypedict.invoke("Please provide the details of the movie avengers")
response
```

```python
class Actor(TypedDict):
    name: str
    role: str

class MovieDetails(TypedDict):
    title: str
    year: int
    cast: list[Actor]
    genres: list[str]
    budget: float | None = Field(None, description="Budget in millions USD")

model_with_structure = model.with_structured_output(MovieDetails)

response = model_with_structure.invoke("Provide details about the movie Avengers")
response
```
