# Assignment on Store_sales dataset

**Load the Dataset**

```python
import pandas as pd
f=pd.read_csv('store_sales.csv')
f
```

**Calculate the total sales for each store across all months.**

```python
df=pd.DataFrame(f)
df.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum()
```

**Find the average sales for each month across all stores**

```python
df.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean()
```

**Identify the store with the highest total sales.**

```python
df['Total']=df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum(axis=1)
print(df.loc[df['Total'].idxmax()])
```

```python
city_sales=pd.DataFrame(f)
city_sales['Total']=city_sales[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum(axis=1)
print(city_sales.groupby('city')['Total'].sum())
```

**List stores with total sales greater than 200**

```python
city_sales[city_sales['Total']>200]
```

**Which month had the highest average sales across all stores?**


```python
city_sales[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean().idxmax()
```

**Which city generated the most revenue?**

```python
print(city_sales.groupby('city')['Total'].sum().idxmax())
```
